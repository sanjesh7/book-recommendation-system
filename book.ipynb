## popularity based filtering

```python
##books
```

```python
ratings_with_name=ratings.merge(books,on="ISBN")
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace="true")
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean(numeric_only=True).round(2).reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg-ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df=popular_df[popular_df['num-ratings']>=250].sort_values('avg-ratings',ascending=False).head(100)
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-L',
'num-ratings','avg-ratings']]
```

```python
popular_df
```

# collaborative filtering

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
uwr=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(uwr)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt=final_ratings.pivot_table(index="Book-Title" ,columns="User-ID",values="Book-Rating")
##pt
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
pt.index[704]
```

```python
def recommend(book_name):
    # Clean and normalize book_name
    book_name_cleaned = book_name.lower().replace(" ", "")

    pt_index_cleaned = []

    for name in pt.index:
        cleaned_name = name.lower().replace(" ", "")
        pt_index_cleaned.append(cleaned_name)

    # index fetch
    index = np.where(np.array(pt_index_cleaned) == book_name_cleaned)

    if len(index[0]) == 0:
        return "Book not found"

    index = index[0][0]

    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[0:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))

        data.append(item)

    return data

```

```python
recommend('animal farm')
```
